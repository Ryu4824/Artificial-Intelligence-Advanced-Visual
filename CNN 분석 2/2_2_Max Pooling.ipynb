```python
import numpy as np
def pool2d_mean(img, pool_size=2, stride=2): # 일반적으로 취하는 값. 크기는 절반으로 줄어든다
  height, width = img.shape # 원본 이미지의 높이와 너비
  out_height = int((height - pool_size) / stride + 1) # 출력 이미지의 높이
  out_width = int((width - pool_size) / stride + 1) # 출력 이미지의 너비
  img_ = np.zeros((out_height, out_width), dtype=np.float32) # 소수점 유지 위해 float로 설정

  for y in range(out_height): # 풀링 윈도우가 가능한 모든 위치에서 진행
    for x in range(out_width):
      img_[y, x] = np.mean(img[y*stride:y*stride+pool_size, x*stride:x*stride+pool_size])
  return img_
```

```python
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread("mountain.jpg", cv2.IMREAD_GRAYSCALE)
output = pool2d_mean(img, pool_size=2, stride=2)
cv2_imshow(img)
cv2_imshow(output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
def pool2d_max(img, pool_size=2, stride=2): # 일반적으로 취하는 값
  height, width = img.shape # 원본 이미지의 높이와 너비
  out_height = int((height - pool_size) / stride + 1) # 출력 이미지의 높이
  out_width = int((width - pool_size) / stride + 1) # 출력 이미지의 너비
  img_ = np.zeros((out_height, out_width), dtype=np.uint8) # 결과 이미지 초기화. 원본의 데이터 타입 유지
  for y in range(out_height): # 풀링 윈도우가 가능한 모든 위치에서 진행
    for x in range(out_width):
      img_[y, x] = np.max(img[y*stride:y*stride+pool_size, x*stride:x*stride+pool_size])
  return img_
```

```python
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread("mountain.jpg", cv2.IMREAD_GRAYSCALE)
output = pool2d_max(img, pool_size=2, stride=2)
cv2_imshow(img)
cv2_imshow(output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
