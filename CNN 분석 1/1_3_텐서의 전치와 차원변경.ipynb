```python
x1=np.array([0,1,2])
print(x1)
print("shape:",x1.shape)
print("---------")
x1_1=np.expand_dims(x1,axis=0)
print(x1_1)
print("shape:",x1_1.shape)
print("---------")
x1_2=np.expand_dims(x1,axis=1)
print(x1_2)
print("shape:",x1_2.shape)
print("---------")
```

```python
x1_3=np.array([[0,1,2],[3,4,5]])
print(x1_3)
print("shape:",x1_3.shape)
print("---------")

x1_4=np.expand_dims(x1_3,axis=0)
print(x1_4)
print("shape:",x1_4.shape)
print("---------")

x1_5=np.expand_dims(x1_3,axis=1)
print(x1_5)
print("shape:",x1_5.shape)
print("---------")
```

```python
x2 = np.array([0,1,2])
print(x2)
print("shpae:",x2.shape)

x2_1 = x2[np.newaxis]
print(x2_1)
print("shape:",x2_1.shape)
print("---------")
x2_2 = x2[np.newaxis,:]
print(x2_2)
print("shape:",x2_2.shape)
print("---------")
x2_3 = x2[:,np.newaxis]
print(x2_3)
print("shape:",x2_3.shape)
print("---------")
```

```python
x3 = np.array([[0,1,2],[3,4,5]])
print(x3)
print("shape:",x3.shape)
print("---------")
x3_1 = x3[np.newaxis]
print(x3_1)
print("shape:",x3_1.shape)
print("---------")
x3_2 = x3[np.newaxis,:,:]
print(x3_2)
print("shape:",x3_2.shape)
print("---------")
```

```python
#세번째 축에 차원을 추가로 끼워넣음
x3_3 = x3[:,:,np.newaxis]
print(x3_3)
print("shape:",x3_3.shape)
print("---------")
#두번째 축에 차원을 추가로 끼워넣음
x3_4 = x3[:,np.newaxis,:]
print(x3_4)
print("shape:",x3_4.shape)
print("---------")
```

```python
#VSTACK
n5 = np.arange(1,7).reshape(2,3)
n6 = np.arange(7,16).reshape(3,3)

print(n5)
print(n6)
print("---------")

#열의 사이즈가 같으면 vstack으로 쌓을 수 있다
n7 = np.vstack([n5,n6])
print(n7)

#행렬과 벡터의 연결도 가능하다
n8 = np.arange(1,4)
np.vstack([n7,n8])
```

```python
#HSTACK
n9 = np.arange(1,7).reshape(2,3)
n10 = np.arange(7,15).reshape(2,4)

print(n9)
print(n10)

n11=np.hstack([n9,n10])
print(n11)
print("---------")

n12=np.array(range(1,15)).reshape(2,-1)
print(n12)
print("---------")

np.hstack([n11,n12])
```

```python
#Broadcasting
n13 = np.array([1,2,3,4,5])
print(n13*2)
print(n13+2)

n14 = np.arange(1,10).reshape(3,3)
print(n14)
print("---------")

n15=np.arange(3)
print(n15)
print("---------")
print(n14+n15)

n16=np.arange(3).reshape(3,1)
print(n16)
print(n16.shape)
print("---------")
n17=np.arange(3).reshape(1,3)
print(n17)
print(n17.shape)
print("---------")
print(n16+n17)
```
