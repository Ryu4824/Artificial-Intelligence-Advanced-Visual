```python
data = [5,3,1,4,7,6,5,2]
h=1/3

mavg = [None]
for i in range(1,len(data)-1):
  mavg.append(data[i-1]*h+data[i]*h+data[i+1]*h)
mavg.append(None)

print(mavg)
```

```python
import matplotlib.pyplot as plt

plt.plot(data,color="blue",linestyle="dotted",label="Input Data")
plt.plot(mavg,color="green",linestyle="solid",label="Simple Moving Average")
plt.legend()
plt.show()
```

```python
data = [5,3,1,4,7,6,5,2]
h=[0.25,0.5,0.25]

w_mavg = [None]
for i in range(1,len(data)-1):
  w_mavg.append(data[i-1]*h[0]+data[i]*h[1]+data[i+1]*h[2])
w_mavg.append(None)

print(w_mavg)
```

```python
import matplotlib.pyplot as plt

plt.plot(data,color="blue",linestyle="dotted",label="Input Data")
plt.plot(mavg,color="green",linestyle="solid",label="Simple Moving Average")
plt.plot(w_mavg,color="red",linestyle="dashed",label="Weighted Moving Average")
plt.legend()
plt.show()
```

```python
data = [0,0,0,10,10,10,10,10,10,10,0,0,0]
h=[-1,2,-1]

w_mavg = [None]
for i in range(1,len(data)-1):
  w_mavg.append(data[i-1]*h[0]+data[i]*h[1]+data[i+1]*h[2])
w_mavg.append(None)

print(w_mavg)
```

```python
import matplotlib.pyplot as plt

plt.plot(data,color="blue",label="Input Data",marker='o')
plt.plot(w_mavg,color="red",label="Simple Moving Average",marker='o')
plt.legend()
plt.show()
```
